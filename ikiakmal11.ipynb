```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords') # Ensure stopwords are available

# Ensure stop_words is defined for this cell's execution
stop_words = set(stopwords.words('indonesian'))

def remove_stopWords(text):
  tokens = text.split()
  filtered = []

  #loop setiap kata
  for word in tokens:
    if word not in stop_words:
      filtered.append(word)

  #menggabungkan kembali kata jdi kalimat
  cleaned_text = " ".join(filtered)
  return cleaned_text

contoh = "Saya adalah orang keren"
print(f'original: {contoh}')
print(f'Setelah proses: {remove_stopWords(contoh)}')
```

```python
df['clean'] = df['clean'].apply(remove_stopWords)
df.head()
```

## Tahap TF

```python
def compute_word_frequency(text):
    words = word_tokenize(text)  # Melakukan yokenisasi teks menjadi kata
    frequency_table = {}         # Dictionary yang berguna untuk menyimpan kata dan frekuensinya

    for word in words:           # Iterasi setiap kata dalam list words
        if word in frequency_table: # Jika kata sudah ada di frequency_table tambahkan hitungannya
            frequency_table[word] += 1 # tamnbahkan frekuensi kata
        else:
            frequency_table[word] = 1 # Jika kata belum ada,  tambahkna ke frequency_tabele dengan frekuensi

    return frequency_table
```

```python
df['word_frequency'] = df['clean'].apply(compute_word_frequency) # Menghitung fekuensi stiap kata dalam teks
df.head()
```

## Tahap IDF

```python
# Jumlah total dokumen
N = len(df)

# Menghitung Document Frequency (dari banyak dokumen mengandung term tertentu)
document_frequencies = {}

for word_freq in df['word_frequency']:
    for word in word_freq.keys():
        if word in document_frequencies:
            document_frequencies[word] += 1
        else:
            document_frequencies[word] = 1

# Menyimpan kedalam bentuk dictionary
idf = {}

# Menghitung IDF untuk setiap kata
for word, df_t in document_frequencies.items():
    idf[word] = math.log(1 + (N / (1 + df_t))) # logatrima dari (total dokumen / total dokumen yang mengandung kata)

# Cek beberapa contoh nilai IDF
list(idf.items())[:10]
```

## TF-IDF

Untuk mendapatkan hasil tf-idf hanya perlu melakukan hasil TF * hasil IDF

```python
# Menghitung TF-IDF untuk setiap dokumen
def compute_tf_idf(word_freq, idf_dict):
    tf_idf = {}
    for word, tf in word_freq.items():
        if word in idf_dict:
            # TF-IDF = TF * IDF
            tf_idf[word] = tf * idf_dict[word]
    return tf_idf

tf_idf_list = []

for freq in df['word_frequency']:
    tf_idf_value = compute_tf_idf(freq, idf)
    tf_idf_list.append(tf_idf_value)

df['tf_idf'] = tf_idf_list
df.head()
```

```python
# Pilih dokumen berdasarkan id berita
id_berita = 19  # ubah sesuai id berita yang ingin ditampilkan
row = df[df['id'] == id_berita].iloc[0]

# Ambil data TF dan TF-IDF dari dokumen tersebut
tf_data = row['word_frequency']
tfidf_data = row['tf_idf']

rows = []
for term, tf in tf_data.items():
    idf_value = idf.get(term, 0)
    tfidf_value = tfidf_data.get(term, 0)
    rows.append({
        'term': term,
        'tf': tf,
        'idf': round(idf_value, 4),
        'tfidf': round(tfidf_value, 4)
    })

df_tfidf_detail = pd.DataFrame(rows)

# Mengurutkan dari nilai tfidf terbesar
df_tfidf_detail = df_tfidf_detail.sort_values(by='tfidf', ascending=False)

print(f"Hasil TF-IDF untuk id berita = {id_berita}\n")
print(df_tfidf_detail.to_string(index=False))
```

## Ranked Retrieval

```python
# membuat funsi untuk cosine similarity
def cosine_similarity_manual(vec1, vec2):
    dot_product = np.dot(vec1, vec2)                       # perkalian titik (∑ Ai * Bi)
    norm_a = np.linalg.norm(vec1)                          # panjang vektor A (akar ∑ Ai²)
    norm_b = np.linalg.norm(vec2)                          # panjang vektor B (akar ∑ Bi²)
    if norm_a == 0 or norm_b == 0:                         # untuk menghindari pambagian nol
        return 0
    return dot_product / (norm_a * norm_b)

def preprocess_query(query):
    query = caseFolding(query)
    query = punctuationRemoval(query)
    query = remove_stopWords(query)
    return query
```

```python
# query yang dinginkan
query = "sepak bola"
query_clean = preprocess_query(query)

query_tf = compute_word_frequency(query_clean)
query_tfidf = compute_tf_idf(query_tf, idf)
```

```python
# Mengumpulan kata dalam setiap dokumen
all_terms_set = set()  # menyimpan kedalam bentuk set

for _, row in df.iterrows():
    # mengambil kata TF-IDF dokumen
    terms_in_doc = row['tf_idf'].keys()

    # menambahkan ke kumpulan semua kata
    all_terms_set.update(terms_in_doc)

# menambahkan kata dari query
all_terms_set.update(query_tfidf.keys())

all_terms = sorted(all_terms_set) #mengurutkan
```

```python
# membuat matrik tfidf menggunakan numpy
print("\nMembentuk matriks TF-IDF untuk seluruh dokumen dokumen")
doc_ids = df['id'].astype(int).tolist()

doc_vectors = np.array([
    [row['tf_idf'].get(term, 0) for term in all_terms]
    for _, row in tqdm(df.iterrows(), total=len(df))
])
```

```python
# Vektor untuk query
query_vector = np.array([query_tfidf.get(term, 0) for term in all_terms])
```

```python
# membuat tabel tf-idf
print("\nMembuat tabel TF-IDF")
df_tfidf_table = pd.DataFrame(doc_vectors.T, index=all_terms, columns=[f"d{doc_id}" for doc_id in doc_ids])
df_tfidf_table["q"] = query_vector
df_tfidf_table.index.name = "Term"
```

```python
from IPython.display import display
display(df_tfidf_table.iloc[10200:10300])
```

```python
query_only = df_tfidf_table[
    (df_tfidf_table["q"] > 0) &
    (df_tfidf_table[[col for col in df_tfidf_table.columns if "tfidf(d" in col]] == 0).all(axis=1)
]
display(query_only)
```

```python
# menghitung cosine similarity dengan vektorisasi
print("\nMenghitung cosine similarity")
dot_products = np.dot(doc_vectors, query_vector)
doc_norms = np.linalg.norm(doc_vectors, axis=1)
query_norm = np.linalg.norm(query_vector)
cosine_similarities = dot_products / (doc_norms * query_norm + 1e-9)
```

```python
# Ranking hasil
df_similarity = pd.DataFrame({
    'id_dokumen': doc_ids,
    'similarity': np.round(cosine_similarities, 4)
}).sort_values(by='similarity', ascending=False)

print("\n=== Ranking Hasil Cosine Similarity ===")
print(df_similarity.head(10).to_string(index=False))
```
